### Data Extraction

```python
import requests
import pandas as pd
```

```python
url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=10&page=1&sparkline=false"
```

```python
response = requests.get(url)

print("Status Code:", response.status_code)
```

```python
data = response.json()

print(data)
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
Historical_price = "https://api.coingecko.com/api/v3/coins/bitcoin/market_chart?vs_currency=usd&days=30&interval=daily"
```

```python
response = requests.get(Historical_price)

print("Status Code:", response.status_code)
```

```python
Price = response.json()

print(Price)
```

```python
Price = pd.DataFrame(Price)
```

```python
Price.head()
```

```python
Price.info()
```

```python
Market = "https://api.coingecko.com/api/v3/global"
```

```python
response = requests.get(Market)

print("Status Code:", response.status_code)
```

```python
Market = response.json()

print(Market)
```

```python
Market_summary = pd.DataFrame(Market)
```

### Data Cleaning

```python
df.columns
```

```python
Columns_to_keep = [
    'name', 'symbol','current_price', 'market_cap', 'market_cap_rank', 
    'total_volume', 'high_24h', 'low_24h', 'market_cap_change_percentage_24h',
    'last_updated']
Coin_market = df[Columns_to_keep]
```

```python
Coin_market
```

```python
Coin_market.info()
```

```python
Coin_market.duplicated().sum()
```

```python
Coin_market['current_price'] = (Coin_market['current_price'].round(2))
```

```python
Coin_market['market_cap'] = (Coin_market['market_cap'] / 1_000_000_000).round(2).astype(str) + "B"
```

```python
temp = Price ["total_volumes"].astype(str).str.strip("[]").str.split(",", expand= True)
```

```python
Price["volume_caps"] = temp [0]
```

```python
Price["main_volume"] = temp[temp.columns[-1]]
```

```python
temp2 = Price ["market_caps"].astype(str).str.strip("[]").str.split(",", expand= True)
```

```python
Price["Date_caps"] = temp2 [0]
```

```python
Price["main_caps"] = temp2[temp2.columns[-1]]
```

```python
temp3 = Price ["prices"].astype(str).str.strip("[]").str.split(",", expand= True)
```

```python
Price["price_caps"] = temp3 [0]
```

```python
Price["main_price"] = temp3[temp3.columns[-1]]
```

```python
Cols = ['main_price', 'volume_caps','main_volume','main_caps']

Price = Price [Cols]      
```

```python
Price
```

```python
Price.rename(columns={'main_price':'price',
                      'volume_caps':'date',
                      'main_volume':'total_volume',
                      'main_caps':'Market_cap'}, inplace= True)
```

```python
Price['price'] = pd.to_numeric(Price['price'])
```

```python
Price['date'] = pd.to_numeric(Price['date'])
```

```python
Price['price'] = (Price['price'].round(2))
```

```python
Price['Market_cap'] = pd.to_numeric(Price['Market_cap'])
```

```python
Price['Market_cap'] = (Price['Market_cap'] / 1_000_000_000).round(2).astype(str) + "B"
```

```python
Price['date'] = pd.to_datetime(Price['date'], unit='ms')
```

```python
Price['total_volume'] = pd.to_numeric(Price['total_volume'])
```
